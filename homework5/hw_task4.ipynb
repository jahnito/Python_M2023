```python
import pandas as pd
import numpy as np
import math
```

```python
'''
Загрузите данные из файла wr88125.txt в объект DataFrame, добавьте заголовки к столбцам: «index», «year», «month», «day», «min_t», «average_t», «max_t», «rainfall»
'''
with open('wr88125.txt', 'r') as f, open('data.txt', 'w') as nf:
    for line in f:
        nf.write(line.replace(' ', ''))

names = ['index','year','month','day','min_t','average_t','max_t','rainfall']
df = pd.read_table('data.txt', sep=';', header=None, names=names, encoding='windows-1251')
df
```

```python
'''
Удалите столбец index.
'''
del df['index']
df[155:168]
```

```python
'''
Используя метод info(), oтветьте на вопросы:
	Есть ли в данных пропущенные значения? - ответ Да
	В каком столбце данных больше всего пропущенных - ответ max_t
'''

print(df.info())
```

```python
'''
В данных за какой год больше всего пропусков?
'''

cols = names[4:]

s1 = df['year']
s2 = pd.Series(df[cols].isna().any(axis=1))

df2 = pd.DataFrame()

df2['year'] = s1
df2['na'] = s2

df2[df2['na']].value_counts()
```

```python
'''
Объедините столбцы «Год», «Месяц» и «День» в один столбец «Дата» в формате гггг-мм-дд (2000-01-20). 
Данные в новом столбце должны иметь формат datetime;
'''

s_d = pd.to_datetime(df['year']*10000 + df['month']*100 + df['day'], format='%Y%m%d')

s_d.name = 'date'

df10 = pd.concat([df['year'], s_d, df[cols]], axis=1)
df10
```

```python
df10['date']
```

```python
'''
Для каждого наблюдения рассчитайте размах температур (разность максимальной и минимальной суточных температур) и количество предшествующих ему дней без осадков (используйте циклы Python и условный оператор)
'''

min_t = df10['min_t'].max() - df10['min_t'].min()
max_t = df10['max_t'].max() - df10['max_t'].min()
average_t = df10['average_t'].max() - df10['average_t'].min()

k = None
for i in range(len(df10)):
    if df10['min_t'].max() == df10.iloc[i]['min_t']:
        k = i
        break
    if df10['min_t'].min() == df10.iloc[i]['min_t']:
        k = i
        break

c_min = 0
for i in range(k, 0, -1):
    if df10.iloc[i]['rainfall'] == 0:
        c_min += 1
    else:
        break


k = None
for i in range(len(df10)):
    if df10['max_t'].max() == df10.iloc[i]['max_t']:
        k = i
        break
    if df10['max_t'].min() == df10.iloc[i]['max_t']:
        k = i
        break

c_max = 0
for i in range(k, 0, -1):
    if df10.iloc[i]['rainfall'] == 0:
        c_max += 1
    else:
        break


k = None
for i in range(len(df10)):
    if df10['average_t'].max() == df10.iloc[i]['average_t']:
        k = i
        break
    if df10['average_t'].min() == df10.iloc[i]['average_t']:
        k = i
        break

c_av = 0
for i in range(k, 0, -1):
    if df10.iloc[i]['rainfall'] == 0:
        c_av += 1
    else:
        break

print(f'Размах для минимальной температуры {min_t}, количество дней без осадков {c_min}')
print(f'Размах для максимальной температры {max_t}, количество дней без осадков {c_max}')
print(f'Размах для средней температуры {average_t}, количество дней без осадков {c_av}')
```

```python
'''
Определите самый длинный период засухи.
'''
big_zasuha = 0
c = 0
for i in df['rainfall']:
    if i == 0:
        c += 1
        if c > big_zasuha:
            big_zasuha = c
    else:
        c = 0

big_zasuha
```

```python
'''
Для каждого года вычислите среднегодовую температуру и общее количество осадков. Запишите результаты в объекты Series
'''

result_t = []
result_r = []
for i in df10['year'].unique():
    mean_value_t = df[df['year'] == i]['average_t'].mean()
    mean_value_r = df[df['year'] == i]['rainfall'].mean()
    result_t.append(mean_value_t)
    result_r.append(mean_value_r)
    # print(f'{i} - {round(mean_value_t, 2)} - {round(mean_value_r,2)}')

s1 = pd.Series(result_t, index=df10['year'].unique())
s2 = pd.Series(result_r, index=df10['year'].unique())
print(s1[s1 == s1.max()])
print(s1[s1 == s1.min()])

print(s2[s2 == s2.max()])
print(s2[s2 == s2.min()])
```
