```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (15, 10))
```

```python
def f(x):
    return np.sin(x/5) * np.exp(x/10) + 5 * np.exp(-x/1)

mp = np.vectorize(f)

x = np.linspace(1, 15, 50)
y = mp(x)


plt.grid()
plt.plot(x, y)
plt.xlabel('x', fontsize=8)
plt.ylabel('y', fontsize=8)
plt.show()
```
