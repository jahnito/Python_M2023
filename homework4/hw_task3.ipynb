```python
import numpy as np
import scipy.linalg
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sin(x/5) * np.exp(x/10) + 5 * np.exp((-x)/2)

# Отрисовка основной функции
mp = np.vectorize(f)
x = np.linspace(1, 15, 5000)
y = mp(x)

# Матрица для 2-х точек
kx = [1, 15]
A = np.array([[1, 1], [1, 15]])
# Вектор
b = np.array([f(1), f(15)])
c = scipy.linalg.solve(A, b)
x1 = np.array(kx)
y1 = np.array([c[0] + c[1] * x for x in kx])

# Матрица для 3-х точек
kx2 = [1, 8, 15]
A2 = np.array([[1, 1, 1], [1, 8, 64], [1, 15, 225]])
# Вектор
b2 = np.array([f(1), f(8), f(15)])
c2 = scipy.linalg.solve(A2, b2)
x2 = np.array(kx2)
y2 = np.array([c2[0] + c2[1] * x + c2[2] * x ** 2 for x in kx2])

# Матрица для 4-х точек
kx3 = [1, 4, 10, 15]
A3 = np.array([[1, 1, 1, 1], [1, 4, 16, 64], [1, 10, 100, 1000], [1, 15, 225, 3375]])
# Вектор
b3 = np.array([f(1), f(4), f(10), f(15)])
c3 = scipy.linalg.solve(A3, b3)
xs = [1, 4, 10, 15]
x3 = np.array(kx3)
y3 = np.array([c3[0] + c3[1] * x + c3[2] * x ** 2 + c3[3] * x ** 3 for x in kx3])

plt.grid()
plt.plot(x, y, linestyle='dashed')
plt.plot(x1, y1)
plt.plot(x2, y2)
plt.plot(x3, y3)

plt.xlabel('x', fontsize=8)
plt.ylabel('y', fontsize=8)
plt.show()
```
